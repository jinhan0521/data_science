```python
df = pd.read_csv('Downloads/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell' : 'U.S.',
    'Apple' : 'U.S.',
    'Acer' : 'Taiwan',
    'HP' : 'U.S.',
    'Lenovo' : 'China',
    'Alienware' : 'U.S.',
    'Microsoft' : 'U.S.',
    'Asus' : 'Taiwan'
}
```

```python
df['brand'].map(brand_nation)
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.max()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
#이 데이터프레임에서 제일 위에 있는 중국노트북 하나, 대만노트북하나, 미국노트북하나
```

```python
nation_groups.last()
#가장마지막
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```
