```python
#한 데이터프레임은 야채의 가격
#또 다른 데이터프레임에는 야채의 수량
#합치기위해 merge사용, 4가지방법이있음
#1. Inner join
#2. left outer join
#3. right outer join
#4. full outer join
#Inner join은 두 데이터프레임에서 겹치는 부분만 join
#left outer join은 왼쪽 dataframe에 존재하는 것만 join
#full outer join은 양쪽 dataframe에 존재하는 것 join
```

```python
import pandas as pd
```

```python
price_df = pd.read_csv('Downloads/vegetable_price.csv')
quantity_df = pd.read_csv('Downloads/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on='Product')
#Inner join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='left')
```

```python
pd.merge(price_df, quantity_df, on='Product', how='right')
```

```python
pd.merge(price_df, quantity_df, on='Product', how='outer')
```
