```python
#과소적합되지도 않았고 과적합되지도 않은 모델을 찾고싶다
#과소적합을 방지하는 일만 생각해보자
#과소적합은 모델이 너무 간단해서 데이터의 관계를 잘 학습하지못하는 현상
#과소적합을 방지하는 방법은 충분히 복잡한 모델을 쓰는 것
#일반적인 상황에서는 선형회귀보다는 높은 차원의 다항 회귀를 쓰면 된다
```

```python
    from sklearn.linear_model import LinearRegression
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import mean_squared_error
    from sklearn.preprocessing import PolynomialFeatures

    from math import sqrt

    import numpy as np
    import pandas as pd
```

```python
#대학원 합격 데이터
```

```python
ADMISSION_FILE_PATH = 'Downloads/admission_data.csv'
admission_df = pd.read_csv(ADMISSION_FILE_PATH).drop('Serial No.', axis=1)
```

```python
admission_df.head()
#GRE점수, TOEFL 점수, 학생이지원하는대학의등급, 학업계획서점수(SOP)
#추천서점수(LOR), 학점(CGPA), 마지막으로 합격확률(Chance of Admit)
```

```python
X = admission_df.drop(['Chance of Admit '], axis=1)
X.values
```

```python
X.columns
```

```python
polynomial_transformer = PolynomialFeatures(6)
#6차항 변환기를 정의
polynomial_features = polynomial_transformer.fit_transform(X.values)
features = polynomial_transformer.get_feature_names(X.columns)

polynomial_features
```

```python
features
```

```python
X = pd.DataFrame(polynomial_features, columns=features)
X.head()
```

```python
y = admission_df[['Chance of Admit ']]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
```

```python
y_test_predict = model.predict(X_test)
```

```python
mse = mean_squared_error(y_train, y_train_predict)

print("training set에서의 성능")
print("-----------------------")
print(sqrt(mse))

mse = mean_squared_error(y_test, y_test_predict)

print("test set에서의 성능")
print("-------------------")
print(sqrt(mse))
```
