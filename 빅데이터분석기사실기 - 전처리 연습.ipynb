### 카드이용 데이터

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(DataUrl, index_col=0)
```

```python
#Income_Category의 카테고리를 map함수를 이용하여 다음과 같이 변경하여 newIncome컬럼에 매핑
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}

df['newIncome'] = df['Income_Category'].map(lambda x: dic[x])
df['newIncome']
```

```python
#apply 함수를 사용하려면

df['newIncome'] = df['Income_Category'].apply(lambda x: dic[x])
df['newIncome']
```

```python
#Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라
#(0~9) : 0, 10~19:10, 20~29 : 20 .... 그리고 각 구간의 빈도수를 출력
df['AgeState'] = (df['Customer_Age']//10) * 10
df['AgeState'].value_counts().sort_index()
```

```python
#Education_Level의 값중 Graduate 단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여
#newEduLevel 컬럼의 정의하고 빈도수를 출력하라
df['newEduLevel'] = df['Education_Level'].map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()
```

```python
#Marital_Status 컬럼값이 Married이고 Card_Category 컬럼의 값이 Platinum인 경우 1
#그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라
#newState의 각 값들의 빈도수를 출력

def newState(x):
    if (x['Marital_Status'] == 'Married') & (x['Card_Category'] == 'Platinum'):
        return 1
    else:
        return 0

df['newState'] = df.apply(newState, axis=1)
df['newState'].value_counts()

#map은 시리즈에 주로 사용되고, apply는 시리즈나 데이터프레임에 적용 가능
```

```python
#Gender 컬럼 값 M인 경우, male F인 경우 female로 값을 변경하여 Gender컬럼에 새롭게 정의
#각 value의 빈도 출력해라

def Gender(x):
    if x == 'M':
        return 'male'
    elif x == 'F':
        return 'female'
    
df['Gender'] = df['Gender'].apply(Gender)
df['Gender'].value_counts()
```

### 주가 데이터

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
df
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
#pd.to_datetime()으로 datetime 타입으로 변경
```

```python
#Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력
df['Yr_Mo_Dy'].dt.year.unique()
```

```python
#Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터
#해당경우 값은 100을 빼자

def fix_century(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)
df
```

```python
#년도별 각컬럼의 평균값을 구하여라
df.groupby(df['Yr_Mo_Dy'].dt.year).mean().head()
```

```python
#weekday컬럼을 만들고 요일별로 매핑하라
df['weekday'] = df.Yr_Mo_Dy.dt.weekday
df['weekday'].to_frame()
#weekday는 날짜의 요일을 0~6으로 출력
```

```python
#weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck컬럼을 만들어라
df['WeekCheck'] = df['weekday'].apply(lambda x : 1 if x in [5,6] else 0)
df['WeekCheck'].to_frame()
```

```python
#모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을 경우 뒤에 있는 값으로 대체하라
df = df.fillna(method='ffill').fillna(method='bfill')
df.isna().sum()
```

```python
#년도-워을 기준으로 모든 컬럼의 평균값을 구하여라
df.groupby(df['Yr_Mo_Dy'].dt.to_period('M')).mean().head()
```

```python
#RPT 컬럼의 값을 일자별 기준으로 1차차분하라
df['RPT'].diff()
#diff()는 각 요소별로 이전요소와의 차이를 출력
```

```python
#RPT와 VAL의 컬럼을 일주일간격으로 각각 이동평균한값을 구하여라
df[['RPT', 'VAL']].rolling(7).mean()
#rolling은 Pandas에서 시계열 데이터에 대해 이동 평균(moving average) 및 다양한 이동 통계량을 계산하기 위한 기능을 제공하는 메서드
```

### 서울시 미세먼지 데이터

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
```

```python
df
```

```python
#년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경
#서울시의 제공데이터의 경우 0시가 24시로 표현된다
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]
    
    if hour == '24':
        hour = '00:00:00'
        
        FinalDate = pd.to_datetime(date + " " + hour) + datetime.timedelta(days=1)
        
    else :
        hour = hour+':00:00'
        FinalDate = pd.to_datetime(date + " " + hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df

#timedelta(days=1)을 한 이유는, '2022-01-01:24'의 경우 '2022-01-02'와 같기 때문
```

```python
#일자별 영어요일 이름을 dayName컬럼에 저장하라
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df['dayName']
```

```python
#dayName별 PM10등급으 빈도수를 파악하라
ans = df.groupby(['dayName', 'PM10등급'],as_index=False).size()
ans
```

```python
ans.pivot(index='dayName', columns='PM10등급', values='size').fillna(0)
```

```python
#시간이 연속적으로 존재하며 결측치가 없는지 확인하라
df['(년-월-일:시)'].diff().unique()
#시간을 차분했을 때, 모든 차분값이 동일하면 연속이라 판단
```

```python
#오전 10시와 오후10시(22시)의 PM10의 평균값을 각각 구하여라
df.groupby(df['(년-월-일:시)'].dt.hour).mean().loc[[10,22],'PM10'].to_frame()
```

```python
#날짜 컬럼을 index로 만들어라
df.set_index('(년-월-일:시)',inplace=True, drop=True)
df
```

```python
#데이터를 주단위로 뽑아서 최소, 최대, 평균, 표준표차를 구하여라
df.resample('W').agg(['min', 'max', 'mean', 'std'])
#리샘플링은 기존의 시계열 데이터의 주기를 변경하는 것을 의미
```

```python
import pandas as pd

# 전처리 과정

### 제주도 데이터

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/Jeju.csv'
df = pd.read_csv(DataUrl,encoding='euc-kr')

#데이터 마지막 3개행
df.tail(3)

#수치형 변수 가진 컬럼 출력
df.select_dtypes(exclude=object).columns

#범주형 변수 가진 컬럼 출력
df.select_dtypes(include=object).columns

#각 컬럼의 결측치 숫자 파악
df.isna().sum()

#평균 속도 컬럼의 4분위 범위 값을 구하여라
df['평균 속도'].quantile(0.75) - df['평균 속도'].quantile(0.25)

#읍면동명 컬럼의 유일값 갯수 출력
df['읍면동명'].nunique()

### 식당 데이터

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)

#quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 추출
df[df['quantity'] == 3].head(5).reset_index(drop=False)

df[df['quantity'] == 3].head(5).reset_index(drop=False)

#item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장
df['new_price'] = df['item_price'].str[1:].astype('float')
df['new_price']

#new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 저체 갯수
len(df[df['new_price'] <= 5])

#df의 new_price컬럼 값에 따라 오름차순으로 정리하고 index 초기화
df.sort_values(by='new_price',ascending=True).reset_index(drop=True)

#df의 item_name 컬럼 값 중 Chips 포함하는 경우의 데이터를 출력하라
df[df['item_name'].str.contains('Chips')]

#df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력
df.iloc[:,::2]

#df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 인덱싱하라
df[df['item_name'].isin(['Steak Salad', 'Bowl'])]

#df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터프레임화 한 후,
#item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라
df[df['item_name'].isin(['Steak Salad', 'Bowl'])].drop_duplicates('item_name')

#df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터프레임화 한 후,
#item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라
df[df['item_name'].isin(['Steak Salad', 'Bowl'])].drop_duplicates('item_name', keep='last')

#df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱
df.loc[df['item_name'] == 'Izze', 'item_name'] = 'Fizzy Lizzy'
df

#df의 데이터 중 item_name 값이 N으로 시작하는 데이터 추출
df[df['item_name'].str.startswith('N')].head()

#df의 데이터 중 item_name값의 단어갯수가 15개 이상인 데이터를 인덱싱
df[df['item_name'].str.len() >= 15]

### 뉴욕 에어비엔비 데이터

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head()

#host_name의 빈도수를 구하고 host_name으로 정렬
df['host_name'].value_counts().sort_index()

#데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라
df['host_name'].value_counts().to_frame().sort_values(by='host_name', ascending=False).rename(columns={'host_name' : 'counts'})

#neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
#as_index=False는 그룹화된 열을 인덱스로 사용하지 않음
#groupby는 counts()대신 size()를 사용해야함

#neighbourhood_group의 값에 따른 neighbourhood 컬럼 값중 neighbourhood_group그룹의 최댓값들을 출력
df.groupby(['neighbourhood_group', 'neighbourhood'],as_index=False).size().groupby('neighbourhood_group', as_index=False).max()

#neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하라
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
#여러가지를 구할 때, agg를 사용하자
#df.groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])['price']

#neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 구하여라
df.groupby(['neighbourhood', 'neighbourhood_group']).mean()['price']

#neighbourhood 값과 neighbourhood_group값에 따른 price의 평균을 계층적 indexing없이 구하라
df.groupby(['neighbourhood', 'neighbourhood_group']).mean()['price'].unstack()
#unstack()을 해주어 neighbourhood_group이 column으로 바뀌었다

#neighbourhood_group값에 다른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을기준으로 각 값의 비율을 구하여라
(df.groupby(['neighbourhood_group', 'room_type']).size().unstack())/(df.groupby(['neighbourhood_group', 'room_type']).size().unstack()).sum(axis=1).values.reshape(-1,1)

df.groupby(['neighbourhood_group', 'room_type']).size().unstack().sum(axis=1).values.reshape(-1,1)
#shape(-1,1)은 1차원 배열을 열이 1개인 2차원 배열로 바꿔준다

### 카드이용 데이터

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(DataUrl, index_col=0)

#Income_Category의 카테고리를 map함수를 이용하여 다음과 같이 변경하여 newIncome컬럼에 매핑
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}

df['newIncome'] = df['Income_Category'].map(lambda x: dic[x])
df['newIncome']

#apply 함수를 사용하려면

df['newIncome'] = df['Income_Category'].apply(lambda x: dic[x])
df['newIncome']

#Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라
#(0~9) : 0, 10~19:10, 20~29 : 20 .... 그리고 각 구간의 빈도수를 출력
df['AgeState'] = (df['Customer_Age']//10) * 10
df['AgeState'].value_counts().sort_index()

#Education_Level의 값중 Graduate 단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여
#newEduLevel 컬럼의 정의하고 빈도수를 출력하라
df['newEduLevel'] = df['Education_Level'].map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()

#Marital_Status 컬럼값이 Married이고 Card_Category 컬럼의 값이 Platinum인 경우 1
#그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라
#newState의 각 값들의 빈도수를 출력

def newState(x):
    if (x['Marital_Status'] == 'Married') & (x['Card_Category'] == 'Platinum'):
        return 1
    else:
        return 0

df['newState'] = df.apply(newState, axis=1)
df['newState'].value_counts()

#map은 시리즈에 주로 사용되고, apply는 시리즈나 데이터프레임에 적용 가능

#Gender 컬럼 값 M인 경우, male F인 경우 female로 값을 변경하여 Gender컬럼에 새롭게 정의
#각 value의 빈도 출력해라

def Gender(x):
    if x == 'M':
        return 'male'
    elif x == 'F':
        return 'female'
    
df['Gender'] = df['Gender'].apply(Gender)
df['Gender'].value_counts()

### 주가 데이터

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
df

df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
#pd.to_datetime()으로 datetime 타입으로 변경

#Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력
df['Yr_Mo_Dy'].dt.year.unique()

#Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터
#해당경우 값은 100을 빼자

def fix_century(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)
df

#년도별 각컬럼의 평균값을 구하여라
df.groupby(df['Yr_Mo_Dy'].dt.year).mean().head()

#weekday컬럼을 만들고 요일별로 매핑하라
df['weekday'] = df.Yr_Mo_Dy.dt.weekday
df['weekday'].to_frame()
#weekday는 날짜의 요일을 0~6으로 출력

#weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck컬럼을 만들어라
df['WeekCheck'] = df['weekday'].apply(lambda x : 1 if x in [5,6] else 0)
df['WeekCheck'].to_frame()

#모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을 경우 뒤에 있는 값으로 대체하라
df = df.fillna(method='ffill').fillna(method='bfill')
df.isna().sum()

#년도-워을 기준으로 모든 컬럼의 평균값을 구하여라
df.groupby(df['Yr_Mo_Dy'].dt.to_period('M')).mean().head()

#RPT 컬럼의 값을 일자별 기준으로 1차차분하라
df['RPT'].diff()
#diff()는 각 요소별로 이전요소와의 차이를 출력

#RPT와 VAL의 컬럼을 일주일간격으로 각각 이동평균한값을 구하여라
df[['RPT', 'VAL']].rolling(7).mean()
#rolling은 Pandas에서 시계열 데이터에 대해 이동 평균(moving average) 및 다양한 이동 통계량을 계산하기 위한 기능을 제공하는 메서드

### 서울시 미세먼지 데이터

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')

df

#년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경
#서울시의 제공데이터의 경우 0시가 24시로 표현된다
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]
    
    if hour == '24':
        hour = '00:00:00'
        
        FinalDate = pd.to_datetime(date + " " + hour) + datetime.timedelta(days=1)
        
    else :
        hour = hour+':00:00'
        FinalDate = pd.to_datetime(date + " " + hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df

#timedelta(days=1)을 한 이유는, '2022-01-01:24'의 경우 '2022-01-02'와 같기 때문

#일자별 영어요일 이름을 dayName컬럼에 저장하라
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df['dayName']

#dayName별 PM10등급으 빈도수를 파악하라
ans = df.groupby(['dayName', 'PM10등급'],as_index=False).size()
ans

ans.pivot(index='dayName', columns='PM10등급', values='size').fillna(0)

#시간이 연속적으로 존재하며 결측치가 없는지 확인하라
df['(년-월-일:시)'].diff().unique()
#시간을 차분했을 때, 모든 차분값이 동일하면 연속이라 판단

#오전 10시와 오후10시(22시)의 PM10의 평균값을 각각 구하여라
df.groupby(df['(년-월-일:시)'].dt.hour).mean().loc[[10,22],'PM10'].to_frame()

#날짜 컬럼을 index로 만들어라
df.set_index('(년-월-일:시)',inplace=True, drop=True)
df

#데이터를 주단위로 뽑아서 최소, 최대, 평균, 표준표차를 구하여라
df.resample('W').agg(['min', 'max', 'mean', 'std'])
#리샘플링은 기존의 시계열 데이터의 주기를 변경하는 것을 의미

### 국가별 5세이하 사망비율 통계

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv')

#Indicator을 삭제하고 First Tooltip 컬럼에서 신뢰구간에 해당하는 표현을 지워라
df.drop('Indicator', axis=1, inplace=True)
df['First Tooltip'] = df['First Tooltip'].map(lambda x: float(x.split('[')[0]))
df

#년도가 2015년 이상, Dim1이 Both sexes인 케이스만 추출하고
#나라에 따른 년도별 사망률을 데이터프레임화하라
df[(df['Period'] >= 2015) & (df['Dim1'] == 'Both sexes')].pivot(index='Location', columns='Period', values='First Tooltip')

df.pivot_table(index='Dim1',columns='Period',values='First Tooltip',aggfunc='mean')

### 올림픽 메달리스트 정보 데이터

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/winter.csv')

#한국 올림픽 메달리스트 데이터에서 년도에 다른 medal 갯수를 데이터프레임화하라
df[df['Country'] == 'KOR'].pivot_table(index='Year', columns='Medal', aggfunc='size').fillna(0)

#sport 종류에 따른 성별수를 구하여라
df.pivot_table(index='Sport', columns='Gender', aggfunc='size')

### 국가별 5세 이하 사망비율 통계

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)

df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)

#df1과 df2 데이터를 하나의 데이터프레임으로 합쳐라
pd.concat([df1,df2], axis=0)

df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)

#df3과 df4데이터를 하나의 데이터프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해섬나 고려
pd.concat([df3,df4], join='inner')

#df3과 df4데이터를 하나의 데이터프레임으로 합쳐라. 모든 컬럼을포함하고 결측치는 0으로 대체
pd.concat([df3,df4], join='outer').fillna(0)

df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)

#df5과 df6 데이터를 하나의 데이터프레임으로 merge함수를 이용하여 합쳐라
#Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력
pd.merge(df5,df6,on='Algeria', how='inner')

#df5과 df6 데이터를 하나의 데이터프레임으로 merge함수를 이용하여 합쳐라
#Algeria컬럼을 key로 하고 합집합으로 합쳐라
pd.merge(df5, df6, on='Algeria', how='outer')
```
