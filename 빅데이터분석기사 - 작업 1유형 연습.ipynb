### Ïú†ÌäúÎ∏å Îç∞ÏùºÎ¶¨ Ïù∏Í∏∞ÎèôÏòÅÏÉÅ(ÌïúÍµ≠)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
#Ïù∏Í∏∞ÎèôÏòÅÏÉÅ Ï†úÏûëÌöüÏàòÍ∞Ä ÎßéÏùÄ Ï±ÑÎÑê ÏÉÅÏúÑ 10Í∞úÎ™ÖÏùÑ Ï∂úÎ†•(ÎÇ†Ïßú Í∏∞Ï§Ä, Ï§ëÎ≥µ Ìè¨Ìï®)
df.loc[df['channelId'].isin(df['channelId'].value_counts().head(10).index), 'channelTitle'].unique()
```

```python
#ÎÖºÎûÄÏúºÎ°ú Ïù∏Í∏∞ÎèôÏòÅÏÉÅÏù¥ Îêú ÏºÄÏù¥Ïä§Î•º ÌôïÏù∏ÌïòÍ≥† Ïã∂Îã§.
#dislikesÍ∞Ä likesÏàòÎ≥¥Îã§ ÎÜíÏùÄ ÎèôÏòÅÏÉÅÏùÑ Ï†úÏûëÌïú Ï±ÑÎÑêÏùÑ Î™®Îëê Ï∂úÎ†•
df.loc[df['dislikes'] > df['likes'], 'channelTitle'].unique()
```

```python
#Ï±ÑÎÑêÎ™ÖÏùÑ Î∞îÍæº ÏºÄÏù¥Ïä§Í∞Ä ÏûàÎäîÏßÄ ÌôïÏù∏ÌïòÍ≥† Ïã∂Îã§
#channelIdÏùò Í≤ΩÏö∞ Í≥†Ïú†Í∞íÏù¥ÎØÄÎ°ú Ïù¥Î•º ÌÜµÌï¥ Ï±ÑÎÑêÎ™ÖÏùÑ ÌïúÎ≤àÏù¥ÎùºÎèÑ Î∞îÍæº Ï±ÑÎÑêÏùò Í∞ØÏàòÎ•º Íµ¨ÌïòÏó¨Îùº
change = df[['channelTitle', 'channelId']].drop_duplicates()['channelId'].value_counts()
len(change[change>1])
```

```python
#ÏùºÏöîÏùºÏóê Ïù∏Í∏∞ÏûàÏóàÎçò ÏòÅÏÉÅÎì§ Ï§ë Í∞ÄÏû• ÎßéÏùÄ ÏòÅÏÉÅ Ï¢ÖÎ•ò(categoryId)Îäî Î¨¥ÏóáÏù∏Í∞Ä?
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df.loc[df['trending_date2'].dt.day_name() == 'Sunday', 'categoryId'].value_counts().index[0]
#pd.to_datetime()ÏïàÏóê ÏãúÎ¶¨Ï¶àÎ•º ÎÑ£Ïñ¥Ïïº Ìï®
#day_name()ÏùÄ ÏöîÏùº Ïù¥Î¶ÑÏúºÎ°ú Ï∂úÎ†•Îê®
```

```python
#Í∞Å ÏöîÏùºÎ≥Ñ Ïù∏Í∏∞ ÏòÅÏÉÅÎì§Ïùò categoryIdÎäî Í∞ÅÍ∞Å Î™áÍ∞úÏî©Ïù∏ÏßÄ ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú ÌëúÌòÑÌïòÎùº
group = df.groupby([df['trending_date2'].dt.day_name(), 'categoryId'], as_index=False).size()
group.pivot(index='categoryId', columns='trending_date2')
```

```python
#viewcountÎåÄÎπÑ ÎåìÍ∏ÄÏàòÍ∞Ä Í∞ÄÏû• ÎÜíÏùÄ ÏòÅÏÉÅÏùÑ ÌôïÏù∏ÌïòÎùº
target2 = df[df['view_count'] != 0]
t = target2.copy()
t['ratio'] = (target2['comment_count']/target2['view_count'])
t.sort_values(by='ratio', ascending=False).head(1)['title']
```

```python
#viewcountÎåÄÎπÑ ÎåìÍ∏ÄÏàòÍ∞Ä Í∞ÄÏû• ÎÇÆÏùÄ ÏòÅÏÉÅÏùÑ ÌôïÏù∏ÌïòÎùº(0 Ï†úÏô∏)
t[t['ratio'] > 0].sort_values(by='ratio', ascending=True).head(1)['title']
```

```python
#like ÎåÄÎπÑ displikeÏùò ÏàòÍ∞Ä Í∞ÄÏû• Ï†ÅÏùÄ ÏòÅÏÉÅÏùÄ?
target3 = df[(df['likes'] != 0) & (df['dislikes'] != 0)]
t2 = target3.copy()
t2['ratio'] = (target3['likes']/target3['dislikes'])
t2.sort_values(by='ratio', ascending=False).head(1)['title']
```

```python
#Í∞ÄÏû•ÎßéÏùÄ Ìä∏Î†åÎìú ÏòÅÏÉÅÏùÑ Ï†úÏûëÌïú Ï±ÑÎÑêÏùò Ïù¥Î¶ÑÏùÄ?
df.loc[df['channelId'] == df['channelId'].value_counts().index[0], 'channelTitle'].unique()[0]
```

```python
#20Ìöå(20Ïùº)Ïù¥ÏÉÅ Ïù∏Í∏∞ÎèôÏòÅÏÉÅ Î¶¨Ïä§Ìä∏Ïóê Ìè¨Ìï®Îêú ÎèôÏòÅÏÉÅÏùò Ïà´ÏûêÎäî?
(df[['title', 'channelId']].value_counts() >= 20).sum()
```

### Ïú†ÌäúÎ∏å Í≥µÎ≤îÏª®ÌÖêÏ∏† ÎèôÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞

```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
#Í∞Å Îç∞Ïù¥ÌÑ∞Ïùò 'ct'Ïª¨ÎüºÏùÑ ÏãúÍ∞ÑÏúºÎ°ú Ïù∏ÏãùÌï†Ïàò ÏûàÍ≤å datatypeÏùÑ Î≥ÄÍ≤ΩÌïòÍ≥†
#video Îç∞Ïù¥ÌÑ∞Ïùò videonameÏùò Í∞Å valueÎßàÎã§ Î™áÍ∞úÏùò Îç∞Ïù¥ÌÑ∞Ïî© Í∞ÄÏßÄÍ≥† ÏûàÎäîÏßÄ ÌôïÏù∏ÌïòÎùº
channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])

video['videoname'].value_counts()
```

```python
#ÏàòÏßëÎêú Í∞Å videoÏùò Í∞ÄÏû• ÏµúÏã†Ìôî Îêú ÎÇ†ÏßúÏùò viewcountÍ∞íÏùÑ Ï∂úÎ†•ÌïòÎùº
video.sort_values(by=['videoname' ,'ct']).drop_duplicates('videoname', keep='last')[['viewcnt', 'videoname', 'ct']].reset_index(drop=True)
#keep='first': Ï§ëÎ≥µÎêú Ìñâ Ï§ë Ï≤´ Î≤àÏß∏ ÌñâÏùÑ Ïú†ÏßÄÌïòÍ≥† ÎÇòÎ®∏ÏßÄÎ•º Ï†úÍ±∞Ìï©ÎãàÎã§.
#keep='last': Ï§ëÎ≥µÎêú Ìñâ Ï§ë ÎßàÏßÄÎßâ ÌñâÏùÑ Ïú†ÏßÄÌïòÍ≥† ÎÇòÎ®∏ÏßÄÎ•º Ï†úÍ±∞Ìï©ÎãàÎã§.
#keep=False: Ï§ëÎ≥µÎêú Î™®Îì† ÌñâÏùÑ Ï†úÍ±∞Ìï©ÎãàÎã§.
```

```python
#Channel Îç∞Ïù¥ÌÑ∞ Ï§ë 2021-10-03Ïùº Ïù¥ÌõÑ Í∞Å Ï±ÑÎÑêÏùò Ï≤òÏùå Í∏∞Î°ù ÎêêÎçò Íµ¨ÎèÖÏûê Ïàò(subcnt)Î•º Ï∂úÎ†•ÌïòÎùº
target = channel[channel['ct'] > pd.to_datetime('2021-10-03')].sort_values(by=['channelid', 'ct']).drop_duplicates('channelid', keep='first')
target[['channelname', 'subcnt']].reset_index(drop=True)
```

```python
#Í∞ÅÏ±ÑÎÑêÏùò 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00ÍπåÏßÄ Íµ¨ÎèÖÏûêÏàò(subcnt)Ïùò Ï¶ùÍ∞ÄÎüâÏùÑ Íµ¨ÌïòÏó¨Îùº
Oct = channel[channel['ct'] >= '2021-10-03 03:00:00'].sort_values(by=['channelname', 'ct']).drop_duplicates('channelname', keep='first')
Nov = channel[channel['ct'] <= '2021-11-01 15:00:00'].sort_values(by=['channelname', 'ct']).drop_duplicates('channelname', keep='last')
merge_data = pd.merge(Oct, Nov, on='channelname')
merge_data['del'] = merge_data['subcnt_y'] - merge_data['subcnt_x']
merge_data[['channelname', 'del']]
```

```python
#Í∞Å ÎπÑÎîîÏò§Îäî 10Î∂Ñ Í∞ÑÍ≤©ÏúºÎ°ú Íµ¨ÎèÖÏûêÏàò, Ï¢ãÏïÑÏöî, Ïã´Ïñ¥ÏöîÏàò, ÎåìÍ∏ÄÏàòÍ∞Ä ÏàòÏßëÎêúÍ≤ÉÏúºÎ°ú ÏïåÎ†§Ï°åÎã§
#Í≥µÎ≤î EP1Ïùò ÎπÑÎîîÏò§Ï†ïÎ≥¥ Îç∞Ïù¥ÌÑ∞Ï§ë ÏàòÏßëÍ∞ÑÍ≤©Ïù¥ 5Î∂Ñ Ïù¥Ìïò, 20Î∂Ñ Ïù¥ÏÉÅÏù∏ Îç∞Ïù¥ÌÑ∞Íµ¨Í∞ÑÏùò ÏãúÍ∞ÅÏùÑ Î™®Îëê Ï∂úÎ†•
import datetime

ep_one = video[video['videoname'] == ' Í≥µÎ≤î EP1'].sort_values('ct').reset_index(drop=True)
ep_one[(ep_one['ct'].diff(1) >= datetime.timedelta(minutes=20)) | (ep_one['ct'].diff(1) <= datetime.timedelta(minutes=5))]

ep_one[ep_one.index.isin([720, 721, 722, 723, 1635, 1636, 1637])]
```

```python
#Í∞Å ÏóêÌîºÏÜåÎìúÏùò ÏãúÏûëÎÇ†Ïßú(ÎÖÑ-Ïõî-Ïùº)Î•º ÏóêÌîºÏÜåÎìú Ïù¥Î¶ÑÍ≥º Î¨∂Ïñ¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú ÎßåÎì§Í≥† Ï∂úÎ†•ÌïòÎùº
start_video = video.sort_values(by=['videoname', 'ct']).drop_duplicates('videoname', keep='first').reset_index(drop=True)
start_video['date'] = start_video['ct'].dt.date
start_video[['date', 'videoname']]
```

```python
#Í≥µÎ≤îÏùÄ 21:00Ïóê Í≥µÍ∞úÎêúÎã§. Í≥µÍ∞úÎêú ÎÇ† 21ÏãúÏùò viewcnt, ct, videonameÏúºÎ°ú Íµ¨ÏÑ±Îêú
#Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ viewcntÎ•º ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÏó¨ Ï∂úÎ†•ÌïòÎùº
open_video = video[(video['ct'].dt.hour == 21)].sort_values(by=['videoname', 'ct']).drop_duplicates('videoname', keep='first')
open_video[['videoname', 'viewcnt', 'ct']].sort_values(by='viewcnt', ascending=False).reset_index(drop=True)
```

```python
#video Ï†ïÎ≥¥Ïùò Í∞ÄÏû• ÏµúÍ∑º Îç∞Ïù¥ÌÑ∞Îì§ÏóêÏÑú Í∞Å ÏóêÌîºÏÜåÎìúÏùò Ïã´Ïñ¥Ïöî/Ï¢ãÏïÑÏöî ÎπÑÏú®ÏùÑ ratioÏª¨ÎüºÏùÑ ÎßåÎì§Í≥†
#videoname,ratioÎ°ú Íµ¨ÏÑ±Îêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ratioÎ•º Ïò§Î¶ÑÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨
video['ratio'] = (video.sort_values('ct', ascending=False).head(8)['dislikecnt']) / (video.sort_values('ct', ascending=False).head(8)['likecnt'])
video.sort_values('ct', ascending=False).head(8)[['videoname', 'ratio']].sort_values('ratio')
```

```python
#2021-11-01 00:00:00 ~ 15:00:00ÍπåÏßÄ Í∞Å ÏóêÌîºÏÜåÎìúÎ≥Ñ ÏµúÎåìÍ∞í-ÏµúÏÜüÍ∞í

start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')

target = video[(video['ct'] >= start) & (video['ct'] <= end)].reset_index(drop=True)

target[['videoname', 'viewcnt']].groupby('videoname').max() - target[['videoname', 'viewcnt']].groupby('videoname').min()
```

```python
#Ï§ëÎ≥µÎêòÎäî Îç∞Ïù¥ÌÑ∞Ïùò videonameÍ≥º ÏãúÍ∞ÑÎåÄ
video.loc[video.duplicated(), ['videoname', 'ct']]
```

### ÏõîÎìúÏªµ Ï∂úÏ†ÑÏÑ†Ïàò Í≥®Í∏∞Î°ù Îç∞Ïù¥ÌÑ∞

```python
import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
#ÎÇòÎùºÎ≥Ñ Í≥®ÎìùÏ†êÏàò ÏÉÅÏúÑ 5Í∞ú Íµ≠Í∞ÄÏôÄ ÎìùÏ†êÏàò
df.groupby('Country').sum().sort_values(by='Goals', ascending=False).head(5)
```

```python
#Í≥®ÎìùÏ†êÏùÑ Ìïú ÏÑ†ÏàòÍ∞Ä Í∞ÄÏû• ÎßéÏùÄ ÎÇòÎùº ÏÉÅÏúÑ 5Í∞ú Íµ≠Í∞ÄÏôÄ Í∑∏ ÏÑ†Ïàò Ïà´ÏûêÎ•º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÌòïÏãùÏúºÎ°ú Ï∂úÎ†•
df.groupby('Country').size().sort_values(ascending=False).head(5).to_frame()
```

```python
#Years Ïª¨ÎüºÏùÄ ÎÖÑÎèÑ-ÎÖÑÎèÑ ÌòïÏãùÏúºÎ°ú Íµ¨ÏÑ±ÎêòÏñ¥ÏûàÍ≥†, Í∞Å ÎÖÑÎèÑÎäî 4ÏûêÎ¶¨ Ïà´ÏûêÏù¥Îã§.
#ÎÖÑÎèÑ ÌëúÍ∏∞Í∞Ä 4ÏûêÎ¶¨ Ïà´ÏûêÎ°ú ÏïàÎêú ÏºÄÏù¥Ïä§Í∞Ä Ï°¥Ïû¨ÌïúÎã§. Ìï¥Îãπ Í±¥ÏùÄ Î™áÍ±¥Ïù∏ÏßÄ Ï∂úÎ†•ÌïòÎùº

def checkFour(x):
    for value in x:
        if len(str(value)) != 4:
            return False
    return True

df['check'] = df['Years'].str.split('-').apply(checkFour)

len(df[df['check'] == False])
```

```python
df2 = df[df['check'] == True]
len(df2.index)
```

```python
#ÏõîÎìúÌÖÅ Ï∂úÏ†ÑÌöüÏàòÎ•º ÎÇòÌÉÄÎÇ¥Îäî 'LenCup Ïª¨ÎüºÏùÑ Ï∂îÍ∞ÄÌïòÍ≥† 4Ìöå Ï∂úÏ†ÑÌïú ÏÑ†ÏàòÏùò Ïà´ÏûêÎ•º Íµ¨ÌïòÏó¨Îùº'
df2['LenCup'] = df2['Years'].str.split('-').apply(lambda x: len(x))
df2['LenCup'] = df2['LenCup'].astype('int')
len(df2[df2['LenCup'] == 4])
```

```python
#Yugoslavia Íµ≠Í∞ÄÏùò ÏõîÎìúÏªµ Ï∂úÏ†ÑÌöüÏàòÍ∞Ä 2ÌöåÏù∏ ÏÑ†ÏàòÎì§Ïùò Ïà´ÏûêÎ•º Íµ¨ÌïòÏó¨Îùº
len(df2[(df2['Country'] == 'Yugoslavia') & (df2['LenCup'] == 2)])
```

```python
#2002ÎÖÑÎèÑÏóê Ï∂úÏ†ÑÌïú Ï†ÑÏ≤¥ ÏÑ†ÏàòÎäî Î™áÎ™ÖÏù∏Í∞Ä?
len(df2[df2['Years'].str.contains('2002')])
```

```python
#Ïù¥Î¶ÑÏóê 'carlos' Îã®Ïñ¥Í∞Ä Îì§Ïñ¥Í∞ÄÎäî ÏÑ†ÏàòÏùò Ïà´ÏûêÎäî Î™áÎ™ÖÏù∏Í∞Ä?
len(df2[df2['Player'].str.lower().str.contains('carlos')])
```

```python
#ÏõîÎìúÏªµ Ï∂úÏ†Ñ ÌöüÏàòÍ∞Ä 1ÌöåÎøêÏù∏ ÏÑ†ÏàòÎì§ Ï§ëÏóêÏÑú Í∞ÄÏû• ÎßéÏùÄ ÎìùÏ†êÏùÑ Ïò¨Î†∏Îçò ÏÑ†ÏàòÎäî ÎàÑÍµ¨Ïù∏Í∞Ä?
df2[df2['LenCup'] == 1].sort_values(by='Goals', ascending=False).reset_index(drop=True)['Player'][0]
```

```python
#ÏõîÎìúÏªµ Ï∂úÏ†ÑÌöüÏàòÍ∞Ä 1Ìöå ÎøêÏù∏ ÏÑ†ÏàòÎì§Ïù¥ Í∞ÄÏû• ÎßéÏùÄ Íµ≠Í∞ÄÎäî Ïñ¥ÎîîÏù∏Í∞Ä?
df2[df2['LenCup'] == 1].groupby('Country').size().sort_values(ascending=False).index[0]
```

```python
#Í∞Å ÏùºÏûêÏùò ÏöîÏùºÏùÑ ÌëúÍ∏∞ÌïòÍ≥† day_name Ïª¨ÎüºÏùÑ Ï∂îÍ∞ÄÌïòÍ≥† Ïù¥Î•º Ïù¥Ïö©ÌïòÏó¨
#Í∞Å ÏöîÏùºÎ≥Ñ Ïù¥Ïö© ÌöüÏàòÏùò Ï¥ùÌï©ÏùÑ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏúºÎ°ú Ï∂úÎ†•ÌïòÎùº
df['ÎåÄÏó¨ÏùºÏûê'] = pd.to_datetime(df['ÎåÄÏó¨ÏùºÏûê'])
df['day_name'] = df['ÎåÄÏó¨ÏùºÏûê'].dt.day_name()
df.groupby('day_name').sum()['Ïù¥Ïö©Í±¥Ïàò'].sort_values(ascending=False).to_frame()
```

```python
#Í∞Å ÏöîÏùºÎ≥Ñ Í∞ÄÏû• ÎßéÏù¥ Ïù¥Ïö©Ìïú ÎåÄÏó¨ÏÜåÏùò Ïù¥Ïö©ÌöüÏàòÏôÄ ÎåÄÏó¨ÏÜå Î≤àÌò∏Î•º Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏúºÎ°ú
df.groupby(['day_name', 'ÎåÄÏó¨ÏÜåÎ≤àÌò∏'],as_index=False).size().sort_values(by=['day_name', 'size'], ascending=False).drop_duplicates('day_name', keep='first')
```

```python
#ÎÇòÏù¥ÎåÄÎ≥Ñ ÎåÄÏó¨Íµ¨Î∂Ñ ÏΩîÎìúÏùò (ÏùºÏùºÍ∂å/Ï†ÑÏ≤¥ÌöüÏàò)ÎπÑÏú®ÏùÑ Íµ¨ÌïòÏó¨Îùº
#Í∑∏ÎûòÍ≥† Í∞ÄÏû• ÎÜíÏùÄ ÎπÑÏú®ÏùÑ Í∞ÄÏßÄÎäî ÎÇòÏù¥ÎåÄÎ•º ÌôïÏù∏
daily = df[df['ÎåÄÏó¨Íµ¨Î∂ÑÏΩîÎìú'].isin(['ÏùºÏùºÍ∂å', 'ÏùºÏùºÍ∂å(ÎπÑÌöåÏõê)'])]['Ïó∞Î†πÎåÄÏΩîÎìú'].value_counts().sort_index()
total = df['Ïó∞Î†πÎåÄÏΩîÎìú'].value_counts().sort_index()

print((daily/total).sort_values(ascending=False))
print('max ratio age : ',(daily/total).sort_values(ascending=False).index[0])
```

```python
#zone 1 power consumption Ïª¨ÎüºÏùò valueÍ∞íÏùò ÌÅ¨Í∏∞Í∞Ä 
#zone 2 power consumption Ïª¨ÎüºÏùò valueÍ∞íÏùò ÌÅ¨Í∏∞Î≥¥Îã§ ÌÅ∞ Îç∞Ïù¥ÌÑ∞Îì§Ïùò
#humidityÏùò ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÏó¨Îùº
df[df['Zone 1 Power Consumption'] > df['Zone 2  Power Consumption']].Humidity.mean()
```

### ÏÑ±Ïù∏ Í±¥Í∞ïÍ≤ÄÏßÑ Îç∞Ïù¥ÌÑ∞

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/train.csv")
df.head()
```

```python
#ÏàòÏ∂ïÍ∏∞ÌòàÏïïÍ≥º Ïù¥ÏôÑÍ∏∞ ÌòàÏïïÍ∏∞ ÏàòÏπòÏùò Ï∞®Ïù¥Î•º ÏÉàÎ°úÏö¥ Ïª¨Îüº(‚ÄòÌòàÏïïÏ∞®‚Äô) ÏúºÎ°ú ÏÉùÏÑ±ÌïòÍ≥†,
#Ïó∞Î†πÎåÄ ÏΩîÎìúÎ≥Ñ Í∞Å Í∑∏Î£π Ï§ë ‚ÄòÌòàÏïïÏ∞®‚Äô Ïùò Î∂ÑÏÇ∞Ïù¥ 5Î≤àÏß∏Î°ú ÌÅ∞ Ïó∞Î†πÎåÄ ÏΩîÎìúÎ•º Íµ¨ÌïòÏó¨Îùº
df['ÌòàÏïïÏ∞®'] = df['ÏàòÏ∂ïÍ∏∞ÌòàÏïï'] - df['Ïù¥ÏôÑÍ∏∞ÌòàÏïï']
df.groupby('Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)')['ÌòàÏïïÏ∞®'].var().sort_values().index[-5]
```

```python
"""
ÎπÑÎßåÎèÑÎ•º ÎÇòÌÉÄÎÇ¥Îäî ÏßÄÌëúÏù∏ WHtRÎäî ÌóàÎ¶¨ÎëòÎ†à / ÌÇ§Î°ú ÌëúÌòÑÌïúÎã§. 
ÏùºÎ∞òÏ†ÅÏúºÎ°ú 0.58Ïù¥ÏÉÅÏù¥Î©¥ ÎπÑÎßåÏúºÎ°ú Î∂ÑÎ•òÌïúÎã§. 
Îç∞Ïù¥ÌÑ∞Ï§ë WHtR ÏßÄÌëúÏÉÅ ÎπÑÎßåÏù∏ Ïù∏ÏõêÏùò ÎÇ®/Ïó¨ ÎπÑÏú®ÏùÑ Íµ¨ÌïòÏó¨Îùº
"""
df['WHtR'] = df['ÌóàÎ¶¨ÎëòÎ†à']/df['Ïã†Ïû•(5CmÎã®ÏúÑ)']
fat_w = df[df['WHtR'] >= 0.58].groupby('ÏÑ±Î≥ÑÏΩîÎìú').size()['F']
fat_m = df[df['WHtR'] >= 0.58].groupby('ÏÑ±Î≥ÑÏΩîÎìú').size()['M']
fat_m/fat_w
```

### ÏùòÎ£å ÎπÑÏö© ÏòàÏ∏° Îç∞Ïù¥ÌÑ∞

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
train.head()
```

```python
#Ìù°Ïó∞ÏûêÏôÄ ÎπÑÌù°Ïó∞Ïûê Í∞ÅÍ∞Å chargesÏùò ÏÉÅÏúÑ 10% Í∑∏Î£πÏùò ÌèâÍ∑†Ïùò Ï∞®Ïù¥Îäî?
smoke_90 = train[train['smoker']== 'yes']['charges'].quantile(0.9)
nsmoke_90 = train[train['smoker']== 'no']['charges'].quantile(0.9)

smoke_90_mean = train[(train['smoker'] == 'yes') & (train['charges'] >= smoke_90)]['charges'].mean()
nsmoke_90_mean = train[(train['smoker'] == 'no') & (train['charges'] >= nsmoke_90)]['charges'].mean()

smoke_90_mean - nsmoke_90_mean
```
