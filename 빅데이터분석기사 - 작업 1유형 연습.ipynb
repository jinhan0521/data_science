### 유튜브 데일리 인기동영상(한국)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
#인기동영상 제작횟수가 많은 채널 상위 10개명을 출력(날짜 기준, 중복 포함)
df.loc[df['channelId'].isin(df['channelId'].value_counts().head(10).index), 'channelTitle'].unique()
```

```python
#논란으로 인기동영상이 된 케이스를 확인하고 싶다.
#dislikes가 likes수보다 높은 동영상을 제작한 채널을 모두 출력
df.loc[df['dislikes'] > df['likes'], 'channelTitle'].unique()
```

```python
#채널명을 바꾼 케이스가 있는지 확인하고 싶다
#channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라
change = df[['channelTitle', 'channelId']].drop_duplicates()['channelId'].value_counts()
len(change[change>1])
```

```python
#일요일에 인기있었던 영상들 중 가장 많은 영상 종류(categoryId)는 무엇인가?
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df.loc[df['trending_date2'].dt.day_name() == 'Sunday', 'categoryId'].value_counts().index[0]
#pd.to_datetime()안에 시리즈를 넣어야 함
#day_name()은 요일 이름으로 출력됨
```

```python
#각 요일별 인기 영상들의 categoryId는 각각 몇개씩인지 하나의 데이터프레임으로 표현하라
group = df.groupby([df['trending_date2'].dt.day_name(), 'categoryId'], as_index=False).size()
group.pivot(index='categoryId', columns='trending_date2')
```

```python
#viewcount대비 댓글수가 가장 높은 영상을 확인하라
target2 = df[df['view_count'] != 0]
t = target2.copy()
t['ratio'] = (target2['comment_count']/target2['view_count'])
t.sort_values(by='ratio', ascending=False).head(1)['title']
```

```python
#viewcount대비 댓글수가 가장 낮은 영상을 확인하라(0 제외)
t[t['ratio'] > 0].sort_values(by='ratio', ascending=True).head(1)['title']
```

```python
#like 대비 displike의 수가 가장 적은 영상은?
target3 = df[(df['likes'] != 0) & (df['dislikes'] != 0)]
t2 = target3.copy()
t2['ratio'] = (target3['likes']/target3['dislikes'])
t2.sort_values(by='ratio', ascending=False).head(1)['title']
```

```python
#가장많은 트렌드 영상을 제작한 채널의 이름은?
df.loc[df['channelId'] == df['channelId'].value_counts().index[0], 'channelTitle'].unique()[0]
```

```python
#20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?
(df[['title', 'channelId']].value_counts() >= 20).sum()
```

### 유튜브 공범컨텐츠 동영상 데이터

```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
#각 데이터의 'ct'컬럼을 시간으로 인식할수 있게 datatype을 변경하고
#video 데이터의 videoname의 각 value마다 몇개의 데이터씩 가지고 있는지 확인하라
channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])

video['videoname'].value_counts()
```

```python
#수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라
video.sort_values(by=['videoname' ,'ct']).drop_duplicates('videoname', keep='last')[['viewcnt', 'videoname', 'ct']].reset_index(drop=True)
#keep='first': 중복된 행 중 첫 번째 행을 유지하고 나머지를 제거합니다.
#keep='last': 중복된 행 중 마지막 행을 유지하고 나머지를 제거합니다.
#keep=False: 중복된 모든 행을 제거합니다.
```

```python
#Channel 데이터 중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라
target = channel[channel['ct'] > pd.to_datetime('2021-10-03')].sort_values(by=['channelid', 'ct']).drop_duplicates('channelid', keep='first')
target[['channelname', 'subcnt']].reset_index(drop=True)
```

```python
#각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00까지 구독자수(subcnt)의 증가량을 구하여라
Oct = channel[channel['ct'] >= '2021-10-03 03:00:00'].sort_values(by=['channelname', 'ct']).drop_duplicates('channelname', keep='first')
Nov = channel[channel['ct'] <= '2021-11-01 15:00:00'].sort_values(by=['channelname', 'ct']).drop_duplicates('channelname', keep='last')
merge_data = pd.merge(Oct, Nov, on='channelname')
merge_data['del'] = merge_data['subcnt_y'] - merge_data['subcnt_x']
merge_data[['channelname', 'del']]
```

```python
#각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다
#공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분 이상인 데이터구간의 시각을 모두 출력
import datetime

ep_one = video[video['videoname'] == ' 공범 EP1'].sort_values('ct').reset_index(drop=True)
ep_one[(ep_one['ct'].diff(1) >= datetime.timedelta(minutes=20)) | (ep_one['ct'].diff(1) <= datetime.timedelta(minutes=5))]

ep_one[ep_one.index.isin([720, 721, 722, 723, 1635, 1636, 1637])]
```

```python
#각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터프레임으로 만들고 출력하라
start_video = video.sort_values(by=['videoname', 'ct']).drop_duplicates('videoname', keep='first').reset_index(drop=True)
start_video['date'] = start_video['ct'].dt.date
start_video[['date', 'videoname']]
```

```python
#공범은 21:00에 공개된다. 공개된 날 21시의 viewcnt, ct, videoname으로 구성된
#데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라
open_video = video[(video['ct'].dt.hour == 21)].sort_values(by=['videoname', 'ct']).drop_duplicates('videoname', keep='first')
open_video[['videoname', 'viewcnt', 'ct']].sort_values(by='viewcnt', ascending=False).reset_index(drop=True)
```

```python
#video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio컬럼을 만들고
#videoname,ratio로 구성된 데이터프레임을 ratio를 오름차순으로 정렬
video['ratio'] = (video.sort_values('ct', ascending=False).head(8)['dislikecnt']) / (video.sort_values('ct', ascending=False).head(8)['likecnt'])
video.sort_values('ct', ascending=False).head(8)[['videoname', 'ratio']].sort_values('ratio')
```

```python
#2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 최댓값-최솟값

start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')

target = video[(video['ct'] >= start) & (video['ct'] <= end)].reset_index(drop=True)

target[['videoname', 'viewcnt']].groupby('videoname').max() - target[['videoname', 'viewcnt']].groupby('videoname').min()
```

```python
#중복되는 데이터의 videoname과 시간대
video.loc[video.duplicated(), ['videoname', 'ct']]
```

### 월드컵 출전선수 골기록 데이터

```python
import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
#나라별 골득점수 상위 5개 국가와 득점수
df.groupby('Country').sum().sort_values(by='Goals', ascending=False).head(5)
```

```python
#골득점을 한 선수가 가장 많은 나라 상위 5개 국가와 그 선수 숫자를 데이터프레임형식으로 출력
df.groupby('Country').size().sort_values(ascending=False).head(5).to_frame()
```

```python
#Years 컬럼은 년도-년도 형식으로 구성되어있고, 각 년도는 4자리 숫자이다.
#년도 표기가 4자리 숫자로 안된 케이스가 존재한다. 해당 건은 몇건인지 출력하라

def checkFour(x):
    for value in x:
        if len(str(value)) != 4:
            return False
    return True

df['check'] = df['Years'].str.split('-').apply(checkFour)

len(df[df['check'] == False])
```

```python
df2 = df[df['check'] == True]
len(df2.index)
```

```python
#월드텁 출전횟수를 나타내는 'LenCup 컬럼을 추가하고 4회 출전한 선수의 숫자를 구하여라'
df2['LenCup'] = df2['Years'].str.split('-').apply(lambda x: len(x))
df2['LenCup'] = df2['LenCup'].astype('int')
len(df2[df2['LenCup'] == 4])
```

```python
#Yugoslavia 국가의 월드컵 출전횟수가 2회인 선수들의 숫자를 구하여라
len(df2[(df2['Country'] == 'Yugoslavia') & (df2['LenCup'] == 2)])
```

```python
#2002년도에 출전한 전체 선수는 몇명인가?
len(df2[df2['Years'].str.contains('2002')])
```

```python
#이름에 'carlos' 단어가 들어가는 선수의 숫자는 몇명인가?
len(df2[df2['Player'].str.lower().str.contains('carlos')])
```

```python
#월드컵 출전 횟수가 1회뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는 누구인가?
df2[df2['LenCup'] == 1].sort_values(by='Goals', ascending=False).reset_index(drop=True)['Player'][0]
```

```python
#월드컵 출전횟수가 1회 뿐인 선수들이 가장 많은 국가는 어디인가?
df2[df2['LenCup'] == 1].groupby('Country').size().sort_values(ascending=False).index[0]
```

### 서울시 따릉이 이용정보 데이터

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bicycle/seoul_bi.csv')
df.head()
```

```python
#대여일자별 데이터의 수를 데이터프레임으로 출력하고,
#가장 많은 데이터가 있는 날짜를 출력하라
display(df.groupby('대여일자').size().to_frame())
print(df.groupby('대여일자').size().sort_values().index[-1])
```

```python
#각 일자의 요일을 표기하고 day_name 컬럼을 추가하고 이를 이용하여
#각 요일별 이용 횟수의 총합을 데이터 프레임으로 출력하라
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['day_name'] = df['대여일자'].dt.day_name()
df.groupby('day_name').sum()['이용건수'].sort_values(ascending=False).to_frame()
```

```python
#각 요일별 가장 많이 이용한 대여소의 이용횟수와 대여소 번호를 데이터 프레임으로
df.groupby(['day_name', '대여소번호'],as_index=False).size().sort_values(by=['day_name', 'size'], ascending=False).drop_duplicates('day_name', keep='first')
```

```python
#나이대별 대여구분 코드의 (일일권/전체횟수)비율을 구하여라
#그래고 가장 높은 비율을 가지는 나이대를 확인
daily = df[df['대여구분코드'].isin(['일일권', '일일권(비회원)'])]['연령대코드'].value_counts().sort_index()
total = df['연령대코드'].value_counts().sort_index()

print((daily/total).sort_values(ascending=False))
print('max ratio age : ',(daily/total).sort_values(ascending=False).index[0])
```

```python
#연령대별 평균 이동거리를 구하여라
df.groupby('연령대코드').mean()['이동거리']
```

```python
#연령대 코드가 20대인 데이터 추출, 이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터 추출
#대여일자, 대여소 번호 순서로 내림차순 정렬 후 1행부터 200행까지 탄소량의 평균을
#소수점 3째 자리까지 구하라
tw = df[df.연령대코드 =='20대'].reset_index(drop=True)
tw_mean = tw[tw.이동거리 >= tw.이동거리.mean()].reset_index(drop=True)
tw_mean['탄소량'] =tw_mean['탄소량'].astype('float')
target =tw_mean.sort_values(['대여일자','대여소번호'], ascending=False).reset_index(drop=True).iloc[:200].탄소량
result = round(target.sum()/len(target),3)
print(result)
```

```python
#평일 오전 6시,7시,8시의 대여소별 이용 횟수를 구해서 데이터 프레임 형태로 표현한 후
#각 대여시간별 이용 횟수의 상위 3개 대여소와 이용횟수를 출력하라
target = df[(df['day_name'].isin(['Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Monday'])) & (df['대여시간'].isin([6, 7, 8]))]
result = target.groupby(['대여시간', '대여소번호']).size().to_frame('이용 횟수')
result.sort_values(['대여시간', '이용 횟수'],ascending=False).groupby('대여시간').head(3)
```

```python
#이동거리의 평균이상의 이동거리 값을 가지는 데이터를 추출하여 추출데이터의
#이동거리의 표본표준편차 값을 구하여라
df[df['이동거리'] > df['이동거리'].mean()]['이동거리'].std()
```

```python
#남성과 여성의 이동거리값의 평균값을 구하여라
df['sex'] = df['성별'].map(lambda x: '남' if x in ['M','m'] else '여')
answer = df[['sex','이동거리']].groupby('sex').mean()
display(answer)
```

### 전세계 행복도 지표 데이터

```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/happy2/happiness.csv',encoding='utf-8')
df.head()
```

```python
#2018년 2019년 행복랭킹 10위를 차지한 나라의 행복점수의 평균을 구하여라
df[df['행복랭킹'] == 10].점수.mean()
```

```python
#각년도의 행복랭킹 50위이내의 나라들의 각각의 행복점수 평균을 데이터프레임으로 표현
df[df['행복랭킹'] <= 50].groupby('년도').mean().점수
```

```python
#2018년도 데이터들만 추출하여 행복점수와 부패에 대한 인식에 대한 상관계수를 구하여라
df[df.년도 == 2018][['점수', '부패에 대한인식']].corr().iloc[0,1]
```

```python
#2018년도와 2019년도의 행복랭킹이 변화하지않은 나라명의 수를 구하여라
len(df[['행복랭킹','나라명']]) - len(df[['행복랭킹','나라명']].drop_duplicates())
```

```python
#2019년도 데이터들만 추출하여 각변수간 상관계수를 구하고
#내림차순으로 정렬한 후 상위 5개를 데이터프레임으로 출력하라
#컬럼명은 v1,v2,corr으로 표시하라
a = df[df['년도'] == 2019].corr().unstack().to_frame().reset_index().dropna()
result = a[a[0] != 1].sort_values(0, ascending=False).drop_duplicates(0)
result = result.head(5)
result.columns = ['v1', 'v2', 'corr']
result.reset_index(drop=True)
```

```python
#각 년도별 하위 행복점수의 하위 5개 국가의 평균 행복점수를 구하여라
df.groupby('년도').tail(5).groupby('년도').mean().점수
```

```python
#2019년 데이터를 추출하고 해당데이터의 상대GDP 평균 이상의 나라들과
#평균 이하의 나라들의 행복점수 평균을 각각 구하고 그 차이값을 출력하라
df_2019 = df[df['년도'] == 2019]
under_gdp = df_2019[df_2019['상대GDP'] <= df_2019['상대GDP'].mean()]
over_gdp = df_2019[df_2019['상대GDP'] >= df_2019['상대GDP'].mean()]
over_gdp['점수'].mean() - under_gdp['점수'].mean()
```

```python
#각년도의 부패에 대한 인식을 내림차순 정렬했을 때, 상위 20개국가의 부패에 대한 인식의 평균을 구하여라
df.sort_values(['년도','부패에 대한인식'],ascending=False).groupby('년도').head(20).groupby(['년도']).mean()[['부패에 대한인식']]
```

```python
#2018년, 2019년 모두 기록이 있는 나라들 중 년도별 행복점수가 가장 증가한 나라와 그 증가 수치는?
count = df.나라명.value_counts()
target = count[count>=2].index

multiple = df[df.나라명.isin(target)].reset_index(drop=True)
df_18 = multiple[multiple['년도'] == 2018]
df_19 = multiple[multiple['년도'] == 2019]
df2 = pd.merge(df_18, df_19, how='inner', on='나라명')
df2['점수차이'] = df2['점수_y'] - df2['점수_x']
df2.sort_values(ascending=False, by='점수차이').head(1)[['나라명', '점수차이']]
```

### 지역구 에너지 소비량 데이터

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/consum/Tetuan%20City%20power%20consumption.csv')
df.head()
```

```python
#DateTime컬럼을 통해 각 월별로 몇개의 데이터가 있는지 데이터 프레임으로 구하여라
df['DateTime'] = pd.to_datetime(df['DateTime'])
df['DateTime'].dt.month.value_counts().reset_index(drop=True).to_frame()
```

```python
#3월달의 각 시간대별 온도의 평균들 중 가장 낮은 시간대의 온도를 출력
df[df['DateTime'].dt.month == 3].groupby(df['DateTime'].dt.hour).mean()['Temperature'].min()
```

```python
#zone 1 power consumption 컬럼의 value값의 크기가 
#zone 2 power consumption 컬럼의 value값의 크기보다 큰 데이터들의
#humidity의 평균을 구하여라
df[df['Zone 1 Power Consumption'] > df['Zone 2  Power Consumption']].Humidity.mean()
```

```python
#Temperature의 값이 10미만의 경우 A
#10이상 20미만의 경우 B
#20이상 30미만의 경우 C
#그 외의 경우 D라고 할때, 각 단계의 데이터 숫자를 구하여라
def temperature(x):
    if x < 10:
        return 'A'
    elif (x >= 10) & (x < 20):
        return 'B'
    elif (x >= 20) & (x < 30):
        return 'C'
    else:
        return 'D'

df['Temperature'].map(temperature).value_counts()
```

```python
#6월 데이터중 12시의 Temperature의 표준편차
df[(df['DateTime'].dt.hour == 12) & (df['DateTime'].dt.month == 6)]['Temperature'].std()
```

```python
#6월 데이터중 12시의 Temperature의 분산
df[(df['DateTime'].dt.hour == 12) & (df['DateTime'].dt.month == 6)]['Temperature'].var()
```

```python
#Temperature의 평균이상의 Temperature의 값을 가지는 데이터를 
#Temperature를 기준으로 정렬했을 때 4번째 행의 Humidity 값은?
df[df['Temperature'] >= df['Temperature'].mean()].sort_values('Temperature')['Humidity'].iloc[3]
```

### 포켓몬 정보 데이터

```python
#각 Generation의 Attack으로 내림차순 정렬시 상위 5개 데이터들(30개)의 Attack의 전체 평균을 구하여라
df.sort_values(['Generation', 'Attack'], ascending=False).groupby('Generation').head(5)['Attack'].mean()
```

```python
#한번씩만 존재하는 (Type1, Type2)의 쌍을 각 세대은 각각 몇개씩 가지고 있는가?
target = df[['Type 1', 'Type 2']].value_counts().to_frame()

lst = []
for value in target[target[0] == 1].reset_index().values:
    t1 = value[0]
    t2 = value[1]
    
    sp = df[(df['Type 1'] == t1) & (df['Type 2']==t2)]
    lst.append(sp)
    
pd.concat(lst).reset_index(drop=True).Generation.value_counts().sort_index()
```

### 성인 건강검진 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/train.csv")
df.head()
```

```python
#수축기혈압과 이완기 혈압기 수치의 차이를 새로운 컬럼(‘혈압차’) 으로 생성하고,
#연령대 코드별 각 그룹 중 ‘혈압차’ 의 분산이 5번째로 큰 연령대 코드를 구하여라
df['혈압차'] = df['수축기혈압'] - df['이완기혈압']
df.groupby('연령대코드(5세단위)')['혈압차'].var().sort_values().index[-5]
```

```python
"""
비만도를 나타내는 지표인 WHtR는 허리둘레 / 키로 표현한다. 
일반적으로 0.58이상이면 비만으로 분류한다. 
데이터중 WHtR 지표상 비만인 인원의 남/여 비율을 구하여라
"""
df['WHtR'] = df['허리둘레']/df['신장(5Cm단위)']
fat_w = df[df['WHtR'] >= 0.58].groupby('성별코드').size()['F']
fat_m = df[df['WHtR'] >= 0.58].groupby('성별코드').size()['M']
fat_m/fat_w
```
